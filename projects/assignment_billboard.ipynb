```python
melt_bb=pd.melt(bb,id_vars=['artiste','track','genre','date.entered','date.peaked'], 
            value_name='rank', var_name='week')
melt_bb.sample(10)
```

```python
melted=melt_bb.dropna(thresh=7)
melted.shape
```

```python
melted.sample(5)
```

```python
melted.dtypes
```

```python
melted['date.entered']=pd.to_datetime(melted['date.entered'])
```

```python
melted.dtypes
```

```python
melted['date.peaked']=pd.to_datetime(melted['date.peaked'])
```

```python
melted['week'] = melted['week'].str.extract('(\d+)').astype(int)
```

```python
melted.dtypes
```

```python
melted['entered_to_peak'] = melted['date.peaked'] - melted['date.entered'] 
```

```python
melted.sample(5)
```
